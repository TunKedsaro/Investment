```python
from ibapi.client import EClient
from ibapi.wrapper import EWrapper
from ibapi.contract import Contract
import threading
import time
```

```python
class IBApp(EWrapper, EClient):
    def __init__(self):
        EClient.__init__(self, self)
    # Update id that we use refresh
    def nextValidId(self, orderId):
        # Start after connection
        self.reqMarketDataType(3)  # 3 : use delayed data
        c = Contract()
        c.symbol, c.secType, c.exchange, c.currency = "AAPL", "STK", "SMART", "USD"
        self.reqMktData(1, c, "", False, False, [])

    # Status of data
    def marketDataType(self, reqId, marketDataType):
        types = {
            1:"Live",             # Real time data
            2:"Frozen",           # Close market
            3:"Delayed",          # Delay data 
            4:"Delayed-frozen"    # Fronzen 15 min later
            }
        print(f"Data type for req {reqId}: {types.get(marketDataType, marketDataType)}")

    def tickPrice(self, reqId, tickType, price, attrib):
        print(f"TickPrice. ReqId:{reqId}, type:{tickType}, price:{price}")

def run_loop(app):
    app.run()

app = IBApp()
app.connect("127.0.0.1", 7497, clientId=2)

thread = threading.Thread(target=run_loop, args=(app,),daemon=True)
thread.start()

time.sleep(1)
#app.disconnect()
```
